```python
import pandas as pd
from sklearn.model_selection import train_test_split

```

## Import the data

Import the dataset stored in `'heart.csv'`.

```python
# Import the data

df = pd.read_csv('heart.csv')

# Print the first five rows of the data
df.head()
```

## Split the data

Define `X` and `y` where the latter is the `target` variable. This time, follow best practices and also implement a standard train-test split. Assign 25% to the test set and set the `random_state` to 17. 

```python
# Define X and y
X = df.drop('target', axis=1)
y = df['target']

# Split the data into training and test sets


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=17)
print("Training set class distribution:\n", y_train.value_counts())
print("\nTest set class distribution:\n", y_test.value_counts())
```

## Initial Model - scikit-learn

Use scikit-learn to build the logistic regression model.

Turn off the intercept and set the regularization parameter, `C`, to a ridiculously large number such as 1e16. 

```python
from sklearn.linear_model import LogisticRegression

# Initialize the model
initial_model = LogisticRegression(fit_intercept=False, C=1e16, solver='liblinear')

# Fit the model on the training data
initial_model.fit(X_train, y_train)
```

## Create an ROC Curve for the scikit-learn model

Use both the training and test sets.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

y_train_score = initial_model.decision_function(X_train)
y_test_score = initial_model.decision_function(X_test)

# Compute the ROC curves
train_fpr, train_tpr, train_thresholds = roc_curve(y_train, y_train_score)
test_fpr, test_tpr, test_thresholds = roc_curve(y_test, y_test_score)

# Compute and print the AUC scores
train_auc = auc(train_fpr, train_tpr)
test_auc = auc(test_fpr, test_tpr)
print('Train AUC: {:.2f}'.format(train_auc))
print('Test AUC: {:.2f}'.format(test_auc))

# Plot the ROC curves
plt.figure(figsize=(10, 8))
lw = 2

plt.plot(train_fpr, train_tpr, color='blue', lw=lw, label='Train ROC curve')
plt.plot(test_fpr, test_tpr, color='darkorange', lw=lw, label='Test ROC curve')

plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.yticks([i/20.0 for i in range(21)])
plt.xticks([i/20.0 for i in range(21)])
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=16)
plt.legend(loc='lower right', fontsize=12)
plt.show()
```
